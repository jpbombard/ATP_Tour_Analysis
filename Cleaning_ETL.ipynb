Change data types to appropriate values

```python
cols_to_int = [
    'minutes', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon',
    'w_SvGms', 'w_bpSaved', 'w_bpFaced', 'l_ace', 'l_df', 'l_svpt',
    'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced',
    'winner_rank', 'winner_rank_points', 'loser_rank', 'loser_rank_points',
    'winner_ht', 'loser_ht'
]

df[cols_to_int] = df[cols_to_int].apply(pd.to_numeric, errors='coerce').astype('Int64')
```

```python
cols_to_float= [
    'winner_age', 'loser_age'
]

df[cols_to_float] = df[cols_to_float].apply(pd.to_numeric, errors='coerce').astype('Float64')
```

```python
cols_to_category = [
    'surface', 'tourney_level', 'winner_entry', 'loser_entry', 'winner_hand', 'loser_hand'
]

df[cols_to_category] = df[cols_to_category].astype('category')
df.dtypes
```

```python
df.loc[df['tourney_name'].str.contains('Olympics'), 'tourney_level'] = 'O'
```

```python
df['match_id'] = df['tourney_id'].astype(str) + '_' + df['match_num'].astype(str)
```

```python
df['tourney_date'] = df['tourney_date'].astype(str).str.replace(',', '')
```

```python
df['tourney_date'] = pd.to_datetime(df['tourney_date'], format = '%Y%m%d')
df.head()
```

```python
df['tourney_name'] = df['tourney_name'].replace({'Us Open': 'US Open'})
```

```python
df['year'] = df['tourney_date'].dt.year
df['month'] = df['tourney_date'].dt.month
df.sample(10)
```

## Exploratory Data Analysis

Build win-loss table including winning percentage

```python
wins = df['winner_name'].value_counts()
wins.head()
```

```python
losses = df['loser_name'].value_counts()
losses.head()
```

```python
win_loss = pd.DataFrame({
    'Wins': wins,
    'Losses': losses
}).fillna(0)

win_loss.head()
```

```python
win_loss['Wins'] = win_loss['Wins'].astype(int)
win_loss['Losses'] = win_loss['Losses'].astype(int)
```

```python
win_loss['Matches_Played'] = win_loss['Wins'] + win_loss['Losses']
win_loss.sort_values(by = 'Matches_Played', ascending = False).head(10)
```

```python
win_loss['Win_Pct'] = round(win_loss['Wins'] / win_loss['Matches_Played'], 4) * 100

win_loss[win_loss['Matches_Played'] > 50].sort_values(by = 'Win_Pct', ascending = False).head(20)
```

```python
win_loss.index.name = 'player_name'
win_loss = win_loss.reset_index()
win_loss
```

Write a function that will print any players win loss record and their win percentage

```python
def get_record(player):
    wins = (df['winner_name'] == player).sum()
    losses = (df['loser_name'] == player).sum()
    matches = wins + losses
    win_pct = (wins / matches) if matches > 0 else 0
    return (f"{player}'s record is {wins}-{losses}. His win percentage is {round(win_pct, 2) * 100}% across {matches} matches.")

get_record('Tommy Paul')

```

Split matches table into indivudal player match stats with a 'won' boolean column 

```python
winner_stats = df[[
    'winner_id', 'match_id', 'year', 'month', 'winner_name', 'winner_ht', 'winner_hand','tourney_name', 'tourney_level', 'surface', 'round','minutes',
    'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon',
    'w_SvGms', 'w_bpSaved', 'w_bpFaced',
    'winner_rank', 'winner_rank_points'
]].copy()

winner_stats.columns = [
    'player_id', 'match_id', 'year', 'month','player_name', 'height', 'hand','tourney_name', 'tourney_level','surface', 'round','minutes',
    'ace', 'df', 'svpt', 'first_in', 'first_won', 'second_won',
    'sv_gms', 'bp_saved', 'bp_faced',
    'rank', 'rank_points'
]

winner_stats['won'] = True

loser_stats = df[[
    'loser_id', 'match_id', 'year', 'month', 'loser_name', 'loser_ht', 'loser_hand', 'tourney_name','tourney_level','surface', 'round','minutes',
    'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon',
    'l_SvGms', 'l_bpSaved', 'l_bpFaced',
    'loser_rank', 'loser_rank_points'
]].copy()

loser_stats.columns = winner_stats.columns[:-1]  
loser_stats['won'] = False

player_match_stats = pd.concat([winner_stats, loser_stats], ignore_index=True)
player_match_stats.dropna(inplace = True)
player_match_stats.sample(15)
```

Create a player stats table with average stats for players across all matches

```python
player_stats = player_match_stats.groupby(['player_id','player_name', 'height', 'hand']).agg(
    avg_minutes = ('minutes', 'mean'),
    avg_aces = ('ace', 'mean'),
    avg_df = ('df', 'mean'),
    avg_svpt = ('svpt', 'mean'),
    avg_first_in = ('first_in', 'mean'),
    avg_first_won = ('first_won', 'mean'),
    avg_second_won = ('second_won', 'mean'),
    avg_sv_gms = ('sv_gms', 'mean'),
    avg_bp_saved = ('bp_saved', 'mean'),
    avg_bp_faced = ('bp_faced', 'mean'),
).reset_index()

player_stats.sample(10)
```

```python
player_stats.dtypes
```

```python
player_stats = player_stats.round(2) #round all stats to 2 decimals
player_stats.sample(10)
```

Add win-loss statistics to player stats table to get overall player summary 

```python
player_table = player_stats.merge(win_loss, on = 'player_name', how = 'left')
player_table.sample(10)
```

Add titles won

```python
titles = player_match_stats[(player_match_stats['round'] == 'F') & (player_match_stats['won'] == True)]
titles_won = titles.groupby('player_name').size().reset_index(name = 'titles_won')
titles_won.sort_values(by= 'titles_won', ascending = False)
```

```python
player_table = player_table.merge(titles_won, on = 'player_name', how = 'left')
player_table.sort_values(by = 'titles_won', ascending = False)
```

```python
player_table['titles_won'] = player_table['titles_won'].astype('Int64') #convert titles won from float to int
player_table.sort_values(by = 'titles_won', ascending = False).head(20)
```
